```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import re
```

```python
url = "https://www.flipkart.com/search?q=ac&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="
```

```python
Brands = []
Model=[]
Stars = []
Ton = []
CPrice = []
OPrice = []
PowerUsage = []
RoomSize = []
Ratings = []
Reviews = []
prod = []

for i in range(1,38):
    print(url+str(i))
    urls = url+str(i)
    
    page = requests.get(urls)
    data = BeautifulSoup(page.text)
    
    
    for i in data.find_all("div",class_="_4rR01T"):
        Brands.append(i.text.split()[0])
        
    
    for i in data.find_all("div",class_="_4rR01T"):
        Model.append(i.text)


    
    for i in data.find_all("div",class_="_4rR01T"):
        res = (re.findall(r"(\d)\sStar",i.text))
        if len(res)==0:
            Stars.append(np.nan)
        else:
            Stars.append(res[0])


    
    for i in data.find_all("div",class_="_4rR01T"):
        res = re.findall(r"(\d+(\.\d+)?)\sTon",i.text)
        if len(res)==0:
            Ton.append(np.nan)
        else:
            Ton.append(res[0][0])


    
    CPrice_Temp = []
    mydivs = data.find_all("div",class_="_25b18c")
    for x in mydivs:
        Cspan = x.find("div",class_="_30jeq3 _1_WHN1")
        if Cspan is None:
            CPrice_Temp.append(np.nan)
        else:
            CPrice_Temp.append(Cspan.text)
    CPrice.extend(CPrice_Temp[0:24])


    OPrice_Temp = []
    for x in mydivs:
        Ospan = x.find("div",class_="_3I9_wc _27UcVY")
        if Ospan is None:
            OPrice_Temp.append(np.nan)
        else:
            OPrice_Temp.append(Ospan.text)
    OPrice.extend(OPrice_Temp[0:24])


    divs = data.find_all("div", class_="col col-7-12")
    for div in divs:
        divx = div.find("div", class_="_3LWZlK")
        if divx is None:
            prod.append(np.nan)
        else:
            prod.append(divx.text)
    
    
    for i in data.find_all("ul",class_="_1xgFaf"):
        res = re.findall("\:\s(.+)\skWh",i.text)
        if len(res)==0:
            PowerUsage.append(np.nan)
        else:
            PowerUsage.append(res[0])


    
    for i in data.find_all("ul",class_="_1xgFaf"):
        res = re.findall("Room Size: (.*) sqft",i.text)
        if len(res)==0:
            RoomSize.append(np.nan)
        else:
            RoomSize.append(res[0])


    
    divs = data.find_all("div", class_="col col-7-12")
    for div in divs:
        span = div.find("span", class_="_2_R_DZ")
        if span is None:
            Ratings.append(np.nan)
            Reviews.append(np.nan)
        else:
            Ratings.append(span.text.split()[0])
            Reviews.append(span.text.split("&")[1].split()[0])

            
    
    print(len(Brands))
    print(len(Model))
    print(len(Ton))
    print(len(Stars))
    print(len(Reviews))
    print(len(Ratings))
    print(len(OPrice))
    print(len(CPrice))
    print(len(RoomSize))
    print(len(PowerUsage))
    print(len(prod))
```

```python
df1 = pd.DataFrame({"Brands":Brands,
                   "Model":Model,
                  "BEE_Rating":Stars,
                  "Tons":Ton,
                  "Reviews_Count":Reviews,
                  "Ratings_Count":Ratings,
                  "Cust_Review":prod,
                  "Original_Price":OPrice,
                  "Current_Price":CPrice,
                  "Room_Size":RoomSize,
                  "Annual_Power_Usage":PowerUsage})
df1
```
