```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Cleaned_AC_Data_FInal.csv")
df = df.drop(columns="Unnamed: 0")
dfc = df.copy()
dfc
```

```python
dfc["Discount"] = (((dfc["Original_Price"]-df["Current_Price"])/df["Original_Price"])*100).round(1)
```

```python
dfc
```

```python
dfc["Discount"].describe()
```

```python
res = []
for i in dfc["Discount"]:
    if i>=0 and i<=15:
        res.append("A")
    elif i>15 and i<=30:
        res.append("B")
    elif i>30 and i<=45:
        res.append("C")
    else:
        res.append("D")
```

```python
print(res)
```

```python
dfc["Discount_Band"]=res
dfc
```
