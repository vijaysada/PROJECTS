## Filling the missing values in Annual_Power_Usage column with the mean()

```python
table4 = df[df['Annual_Power_Usage'].isna()]
table4
```

```python
table5 = pd.pivot_table(df,index="Brands",columns=["Tons","BEE_Rating"],values="Annual_Power_Usage",aggfunc="mean")
table5
```

```python
table5[1.5][4.0]["Blue"]
```

```python
b_x = []
for i in table4["Brands"]:
    b_x.append(i)
b_x
```

```python
len(b_x)
```

```python
r_x = []
for j in table4["BEE_Rating"]:
    r_x.append(j)
r_x
```

```python
t_x = []
for k in table4["Tons"]:
    t_x.append(k)
t_x
```

```python
ann = []
for m,n,o in zip(t_x,r_x,b_x):
    ann.append(table5[m][n][o].round(2))
ann
```

```python
t4_ind = table4.index
t4_ind
```

```python
table4
```

```python
df["Annual_Power_Usage"][t4_ind]
```

```python
for i in range(len(t4_ind)):
    df.at[t4_ind[i],"Annual_Power_Usage"] = ann[i]
```

```python
df["Annual_Power_Usage"][t4_ind]
```

```python
df.isna().sum()
```

```python
table5[2.0][[3.0,4.0,5.0]]
```

```python
table6 = df[df["Annual_Power_Usage"].isna()]
table6
```

```python
df[(df["Brands"]=="Voltas")&(df["Tons"]==2.0)].groupby("BEE_Rating")["Annual_Power_Usage"].mean()
```

```python
r3 = df[(df["Brands"]=="Voltas")&(df["Tons"]==2.0)].groupby("BEE_Rating")["Annual_Power_Usage"].mean()[3.0]
r3
```

```python
r5 = df[(df["Brands"]=="Voltas")&(df["Tons"]==2.0)].groupby("BEE_Rating")["Annual_Power_Usage"].mean()[5.0]
r5
```

```python
r4 = ((r3+r5)/2).round(2)
r4
```

```python
t6_ind = table6.index
t6_ind
```

```python
df["Annual_Power_Usage"][t6_ind]
```

```python
for i in range(len(t6_ind)):
    df.at[t6_ind[i],"Annual_Power_Usage"]=r4
```
